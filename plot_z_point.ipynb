```python
d = pd.read_csv("./output/cifar100/z_ass/d_euclidean_partition_noniid-#label2_nclient100.npy", header = None)
d_iid = pd.read_csv("./output/cifar100/z_ass/d_euclidean_partition_iid-diff-quantity0.5_nclient100.npy", header = None)

d_noniid_mnist = pd.read_csv("./output/mnist/z_ass/d_euclidean_partition_noniid-#label2_nclient100.npy", header = None)
d_iid_mnist = pd.read_csv("./output/mnist/z_ass/d_euclidean_partition_iid-diff-quantity0.5_nclient100.npy", header = None)
d_noniid_mnist_thuy = pd.read_csv("./z_ass/MNIST_M100/d_euclidean_CNN_P1.6m_B10_L20_non_iid_zipfz0.7", header = None)
d_noniid_mnist_82 = pd.read_csv("./output/mnist/z_ass/d_euclidean_MLP2_G10_partition_noniid82_nclient100.npy", header = None)
d_noniid_mnist_82_label2 = pd.read_csv("./output/mnist/z_ass/d_euclidean_partition_noniid82_#label2_nclient100.npy", header = None)
d_noniid_mnist_label3 = pd.read_csv("./output/mnist/z_ass/d_euclidean_MLP2_G1_partition_noniid-#label3_nclient600.npy", header = None)
```

```python
n= d.shape[0]
d = d.to_numpy()
d_plot = []
for i in range(n):
    for j in range( i+ 1, n):
        d_plot.append(d[i][j])

n= d_iid.shape[0]
d_iid = d_iid.to_numpy()
d_iid_plot = []
for i in range(n):
    for j in range( i+ 1, n):
        d_iid_plot.append(d_iid[i][j])

        
n= d_iid_mnist.shape[0]
d_iid_mnist = d_iid_mnist.to_numpy()
d_iid_mnist_plot = []
for i in range(n):
    for j in range( i+ 1, n):
        d_iid_mnist_plot.append(d_iid_mnist[i][j])

n= d_noniid_mnist.shape[0]
d_noniid_mnist = d_noniid_mnist.to_numpy()
d_noniid_mnist_plot = []
for i in range(n):
    for j in range( i+ 1, n):
        d_noniid_mnist_plot.append(d_noniid_mnist[i][j])

n= d_noniid_mnist_thuy.shape[0]
d_noniid_mnist_thuy = d_noniid_mnist_thuy.to_numpy()
d_noniid_mnist_thuy_plot = []
for i in range(n):
    for j in range( i+ 1, n):
        d_noniid_mnist_thuy_plot.append(d_noniid_mnist_thuy[i][j])

n= d_noniid_mnist_82.shape[0]
d_noniid_mnist_82 = d_noniid_mnist_82.to_numpy()
d_noniid_mnist_82_plot = []
for i in range(n):
    for j in range( i+ 1, n):
        d_noniid_mnist_82_plot.append(d_noniid_mnist_82[i][j])

n= d_noniid_mnist_82_label2.shape[0]
d_noniid_mnist_82_label2 = d_noniid_mnist_82_label2.to_numpy()
d_noniid_mnist_82_label2_plot = []
for i in range(n):
    for j in range( i+ 1, n):
        d_noniid_mnist_82_label2_plot.append(d_noniid_mnist_82_label2[i][j])

n= d_noniid_mnist_label3.shape[0]
d_noniid_mnist_label3 = d_noniid_mnist_label3.to_numpy()
d_noniid_mnist_label3_plot = []
for i in range(n):
    for j in range( i+ 1, n):
        d_noniid_mnist_label3_plot.append(d_noniid_mnist_label3[i][j])
```

```python

fig, axs = plt.subplots(6, 2, figsize=(20,15))

axs[0,0].hist(d_iid_plot, bins=20)
axs[0,0].set_title("histogram: d of iid cifar100")
axs[1,0].hist(d_plot, bins=20)
axs[1,0].set_title("histogram: d of non_iid cifar100")

axs[0,1].hist(d_iid_mnist_plot, bins=20)
axs[0,1].set_title("histogram: d of iid mnist")

axs[1,1].hist(d_noniid_mnist_plot, bins=20)
axs[1,1].set_title("histogram: d of non-iid mnist")

axs[2,1].hist(d_noniid_mnist_thuy_plot, bins=20)
axs[2,1].set_title("histogram: d of zipfz0.7 non-iid mnist")

axs[3,1].hist(d_noniid_mnist_82_plot, bins=20)
axs[3,1].set_title("histogram: d dominated_class82 non-iid mnist")

axs[4,1].hist(d_noniid_mnist_82_label2_plot, bins=20)
axs[4,1].set_title("histogram: d dominated_class82_label2 non-iid mnist")

axs[5,1].hist(d_noniid_mnist_label3_plot, bins=20)
axs[5,1].set_title("histogram: d non-iid mnist_label3")


plt.show()
```
