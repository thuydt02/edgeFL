```python
n_files = len(files)
dfs = []
for i in range(n_files):
    dfs.append(pd.read_csv(path_file + files[i]))
```

```python
fig, ax = plt.subplots(figsize=(15,7))
for i in range(n_files):
    ax.plot(dfs[i].global_round, dfs[i].test_acc, label = files[i])
ax.set_xlabel('Global round')
ax.set_ylabel('Test Accuracy (%)')
ax.legend(loc = 'best')
plt.show()
```
