```python
n= d.shape[0]
d = d.to_numpy()
d_plot = []
for i in range(n):
    for j in range( i+ 1, n):
        d_plot.append(d[i][j])
```

```python
#wPoint# = wPoint[(wPoint.pc1 < 400) & (wPoint.pc2 < 500)]

fig, axs = plt.subplots(1, 2, figsize=(20,8))

axs[0].scatter(wPoint['pc1'],wPoint['pc2'] , s= 20)
axs[0].set_title(point_file)
axs[1].hist(d_plot, bins=20 )
axs[1].set_title(d_file)
plt.show()
```
