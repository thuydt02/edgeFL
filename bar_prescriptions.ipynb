```python
import numpy as np
import seaborn as sns
import plotly.graph_objects as go
import matplotlib.pyplot as plt # For showing plots
from matplotlib.dates import date2num, DayLocator, DateFormatter
import matplotlib.dates as mdates
from matplotlib.font_manager import FontProperties
from matplotlib import cm
```

```python
labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
primary2019 = [8, 8, 7.6, 7.7, 8.1, 7.4]
primary2020 = [8.1, 8.3, 37.4, 29.3, 10.4, 9.5]

nonroutine2019 = [11.6, 11.6, 11.6, 11.4, 12.4, 11.8]
nonroutine2020 = [10.8, 15.1, 82.5, 54.2, 18.5, 16.5]

x = np.arange(len(labels))  # the label locations
width = 0.425  # the width of the bars

fig, ax = plt.subplots(figsize=(10,4.5), ncols=2)
cmap = cm.get_cmap("summer")
color=['#17becf', '#7f7f7f', '#bcbd22']
rects1 = ax[0].bar(x - width/2, primary2019, width, label='2019', color = color[1])
rects2 = ax[0].bar(x + width/2, primary2020, width, label='2020', color = color[0])

rects3 = ax[1].bar(x - width/2, nonroutine2019, width, label='2019', color = color[1])
rects4 = ax[1].bar(x + width/2, nonroutine2020, width, label='2020', color = color[0])

# Add some text for labels, title and custom x-axis tick labels, etc.
ax[0].set_ylabel('New prescriptions from Primary care (in %)',  size = 12)
ax[0].set_xticks(x)
ax[0].set_xticklabels(labels, size = 12)
ax[0].legend()

ax[1].set_ylabel('New nonroutine prescriptions( in %)', size = 12)
ax[1].set_xticks(x)
ax[1].set_xticklabels(labels, size = 12)
ax[1].legend()

for i in range(2):

    right_side = ax[i].spines["right"]
    right_side.set_visible(False)

    left_side = ax[i].spines["left"]
    left_side.set_visible(False)

    top_side = ax[i].spines["top"]
    top_side.set_visible(False)
    ax[i].grid(True, which='major', axis='y')

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
