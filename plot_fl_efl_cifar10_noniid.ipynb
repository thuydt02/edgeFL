```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
path_file = "./output/cifar10/acc/"
```

```python
#L10E2
efl_files = [
#10    
         "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G5000_g_nw_d_euclidean_CNN2_G10_partition_noniid90_nclient300.npy.part.10.weight.pth.acc.csv",     
#20
         "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G5000_g_nw_d_euclidean_CNN2_G10_partition_noniid90_nclient300.npy.part.20.weight.pth.acc.csv",    
#30
         "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G5000_g_nw_d_euclidean_CNN2_G10_partition_noniid90_nclient300.npy.part.30.weight.pth.acc.csv"
]

fl_files = [
        "FL_CNN2_dr10lr0.15_dc0.99_B20_L10_G5000_partition_noniid90_nclient300.npy.weight.pth.acc.csv",
        "FL_CNN2_dr10lr0.15_dc0.99_B20_L15_G5010_partition_noniid90_nclient300.npy.weight.pth.acc.csv",
        "FL_CNN2_dr10lr0.15_dc0.99_B20_L20_G5000_partition_noniid90_nclient300.npy.weight.pth.acc.csv"
]



efl_dfs, fl_dfs = [], []
n_efl_files, n_fl_files = len(efl_files), len(fl_files)

for i in range(n_efl_files):
    efl_dfs.append(pd.read_csv(path_file + efl_files[i]))

for i in range(n_fl_files):
    df = pd.read_csv(path_file + fl_files[i])
    count_nan_in_df = df.isnull().sum().sum()
    print(df.shape)
    print(fl_files[i])
    print("#NAN: ", count_nan_in_df)
    if count_nan_in_df == 0:
        fl_dfs.append(df)
    

features = ['test_acc']

fig, ax = plt.subplots(figsize=(10,7))

for i in range(n_efl_files):
    #df_plot = efl_dfs[i].global_round <= 3000]
    ax.plot(efl_dfs[i].global_round, efl_dfs[i][features[0]], label = efl_files[i])

for i in range(len(fl_dfs)):
    #df_plot = efl_dfs[i].global_round <= 3000]
    ax.plot(fl_dfs[i].global_round, fl_dfs[i][features[0]], label = fl_files[i])

ax.set_title("FL and eFL in nonIID: L10E2 in comparison")
ax.set_ylabel("Test Accuracy (%)")
ax.set_xlabel("Training Time")
ax.legend(loc='best')


plt.show()

```
