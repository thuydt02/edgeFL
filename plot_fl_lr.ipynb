```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
path_file = './output/cifar10/acc/'
fl_files = [
    "FL_CNN2_dr10lr0.01_dc0.99_B20_L10_G5000_partition_noniid90_nclient300.npy.weight.pth.acc.csv",
    "FL_CNN2_dr10lr0.05_dc0.99_B20_L10_G5000_partition_noniid90_nclient300.npy.weight.pth.acc.csv",
    "FL_CNN2_dr10lr0.1_dc0.99_B20_L10_G5000_partition_noniid90_nclient300.npy.weight.pth.acc.csv",
    "FL_CNN2_dr10lr0.15_dc0.99_B20_L10_G5000_partition_noniid90_nclient300.npy.weight.pth.acc.csv",
    "FL_CNN2_dr10lr0.2_dc0.99_B20_L10_G5000_partition_noniid90_nclient300.npy.weight.pth.acc.csv",

    "FL_CNN2_dr10lr0.01_dc0.995_B20_L10_G5000_partition_noniid90_nclient300.npy.weight.pth.acc.csv",
    "FL_CNN2_dr10lr0.05_dc0.995_B20_L10_G5000_partition_noniid90_nclient300.npy.weight.pth.acc.csv",
    "FL_CNN2_dr10lr0.1_dc0.995_B20_L10_G5000_partition_noniid90_nclient300.npy.weight.pth.acc.csv",
    "FL_CNN2_dr10lr0.15_dc0.995_B20_L10_G5000_partition_noniid90_nclient300.npy.weight.pth.acc.csv",
    "FL_CNN2_dr10lr0.2_dc0.995_B20_L10_G5000_partition_noniid90_nclient300.npy.weight.pth.acc.csv"

    
]
```

```python
fl_dfs = []
n_fls = len(fl_files)

for i in range(n_fls):
    fl_dfs.append(pd.read_csv(path_file + fl_files[i]))
```

```python
fig, ax = plt.subplots( figsize=(15,10))
for i in range(n_fls):
    fl_dfs[i].plot(x = 'global_round', y = 'train_loss', ax = ax, kind = 'line', label = fl_files[i])
ax.set_title("FL in lr comparison")
ax.set_xlabel('Global Time')
ax.set_ylabel('Train Loss')
ax.legend(loc='best')
plt.show()
```

```python
fig, ax = plt.subplots( figsize=(15,10))
for i in range(n_fls):
    fl_dfs[i].plot(x = 'global_round', y = 'test_acc', ax = ax, kind = 'line', label = fl_files[i])

ax.set_title("FL in lr comparison")
ax.set_xlabel('Global Time')
ax.set_ylabel('Test Acc')
ax.legend(loc='best')
plt.show()
```
