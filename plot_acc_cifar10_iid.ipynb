```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
path_file = "./output/cifar10/acc/"
```

```python
#L10E2
files = [
#10    
        ["eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G5000_avg5z_rnd_num_client300.part10_partition_iid_nclient300.npy.weight.pth.acc.csv",
         "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G3000_g_nw_d_euclidean_CNN2_G100_partition_iid_nclient300.npy.part.10.weight.pth.acc.csv",
         ],
#20
        ["eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G5000_avg10z_rnd_num_client300.part20_partition_iid_nclient300.npy.weight.pth.acc.csv",
         "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G3000_g_nw_d_euclidean_CNN2_G100_partition_iid_nclient300.npy.part.20.weight.pth.acc.csv",
         ],
    
#30
        ["eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G5000_avg10z_rnd_num_client300.part30_partition_iid_nclient300.npy.weight.pth.acc.csv",
         "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G3000_g_nw_d_euclidean_CNN2_G100_partition_iid_nclient300.npy.part.30.weight.pth.acc.csv",
         ]
]

clusters = ['10', '20', '30']
algs = ['z_rnd', 'z_metis']


n_clusters = len(clusters)
n_algs = len(algs)
dfs = [[] for _ in range(n_clusters)]

for i in range(n_clusters):
    for j in range(n_algs):
        dfs[i].append(pd.read_csv(path_file + files[i][j]))
        

features = ['test_acc']

fig, ax = plt.subplots(int (n_clusters/2) + 1, 2, figsize=(12,10))

for i in range(n_clusters):
    for j in range(n_algs):
        df_plot = dfs[i][j][dfs[i][j].global_round <= 3000]
        ax[int(i/2), i % 2].plot(df_plot.global_round, df_plot[features[0]], label = files[i][j])
    
    ax[int(i/2), i % 2].set_title("cluster_" + str(clusters[i]))
    ax[int(i/2), i % 2].set_ylabel("Test Accuracy (%)")
    ax[int(i/2), i % 2].set_xlabel("Training Time")
    ax[int(i/2), i % 2].legend(loc='best')


plt.show()

```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,6))


for j in range(n_algs):
    for i in range(n_clusters):
        df_plot = dfs[i][j][dfs[i][j].global_round <= 3000]
        ax[j].plot(df_plot.global_round, df_plot[features[0]], label = files[i][j])
    
    ax[j].set_title("alg_" + str(algs[j]))
    ax[j].set_ylabel("Test Accuracy")
    ax[j].set_xlabel("Training Time")
    ax[j].legend(loc='best')

```
