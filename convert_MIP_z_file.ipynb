```python
path = "./output/mnist/z_ass/"

z_mip_file = "MIP_z_n300part10.csv"
new_z_file_name = "z_MIP_n300part10.csv"
dfz = pd.read_csv(path + z_mip_file, header = None)
dfz.head(20)
```

```python
n = dfz.shape[1]
k = dfz.shape[0]
z = np.zeros((n, 1), dtype = int)
a = dfz.values
for j in range(n):
    for i in range(k):
        if a[i,j] > 0:
             z[j,0] = i


print(max(z))
print(min(z))
```

```python
df1 = pd.DataFrame()
df1["z"] = np.squeeze(z)
print(df1.shape)
```
