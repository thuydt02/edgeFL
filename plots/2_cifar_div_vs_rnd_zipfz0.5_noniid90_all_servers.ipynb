```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
iidL10E2_path_file = "./../output/cifar10/acc/iid/L10E2/"
iidL20E5_path_file = "./../output/cifar10/acc/iid/L20E5/"
noniidL10E2_path_file = "./../output/cifar10/acc/noniid/L10E2/"
noniidL20E5_path_file = "./../output/cifar10/acc/noniid/L20E5/"
fontsize = 20
```

```python
#reading files
#div-label2
noniidL10E2_metisp1_files = [
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G3000_g_nw_d_minkowski_p1.0_CNN2_G100_partition_zipfz0.5_noniid90_nclient300.npy.part.10.weight.pth.acc.csv",
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G3000_g_nw_d_minkowski_p1.0_CNN2_G100_partition_zipfz0.5_noniid90_nclient300.npy.part.20.weight.pth.acc.csv",
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G3000_g_nw_d_minkowski_p1.0_CNN2_G100_partition_zipfz0.5_noniid90_nclient300.npy.part.30.weight.pth.acc.csv"
]
#rnd-lable2
noniidL10E2_rnd_files = [
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G3000_1z_rnd_num_client300.part10_partition_zipfz0.5_noniid90_nclient300.npy.weight.pth.acc.csv",
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G3000_1z_rnd_num_client300.part20_partition_zipfz0.5_noniid90_nclient300.npy.weight.pth.acc.csv",
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G3000_1z_rnd_num_client300.part30_partition_zipfz0.5_noniid90_nclient300.npy.weight.pth.acc.csv"

]
#div-label2
noniidL20E5_metisp1_files = [
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L20_E5_G3000_g_nw_d_minkowski_p1.0_CNN2_G100_partition_zipfz0.5_noniid90_nclient300.npy.part.10.weight.pth.acc.csv",
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L20_E5_G3000_g_nw_d_minkowski_p1.0_CNN2_G100_partition_zipfz0.5_noniid90_nclient300.npy.part.20.weight.pth.acc.csv",
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L20_E5_G3000_g_nw_d_minkowski_p1.0_CNN2_G100_partition_zipfz0.5_noniid90_nclient300.npy.part.30.weight.pth.acc.csv"
]
#rnd-label2
noniidL20E5_rnd_files = [
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L20_E5_G3000_1z_rnd_num_client300.part10_partition_zipfz0.5_noniid90_nclient300.npy.weight.pth.acc.csv",
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L20_E5_G3000_1z_rnd_num_client300.part20_partition_zipfz0.5_noniid90_nclient300.npy.weight.pth.acc.csv",
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L20_E5_G3000_1z_rnd_num_client300.part30_partition_zipfz0.5_noniid90_nclient300.npy.weight.pth.acc.csv"
]



noniidL10E2_dfs = []
noniidL10E2_n_files = len(noniidL10E2_metisp1_files)
for i in range(noniidL10E2_n_files):
    metisp1_df = pd.read_csv(noniidL10E2_path_file + noniidL10E2_metisp1_files[i])
    rnd_df = pd.read_csv(noniidL10E2_path_file + noniidL10E2_rnd_files[i])
    df = pd.DataFrame(metisp1_df.global_round)
    df['train_loss'] = rnd_df.train_loss/metisp1_df.train_loss
    df['train_acc'] = rnd_df.train_acc/metisp1_df.train_acc
    df['test_loss'] = rnd_df.test_loss/metisp1_df.test_loss
    df['test_acc'] = rnd_df.test_acc/metisp1_df.test_acc
    noniidL10E2_dfs.append(df)

noniidL20E5_dfs = []
noniidL20E5_n_files = len(noniidL20E5_metisp1_files)
for i in range(noniidL20E5_n_files):
    metisp1_df = pd.read_csv(noniidL20E5_path_file + noniidL20E5_metisp1_files[i])
    rnd_df = pd.read_csv(noniidL20E5_path_file + noniidL20E5_rnd_files[i])
    df = pd.DataFrame(metisp1_df.global_round)
    df['train_loss'] = rnd_df.train_loss/metisp1_df.train_loss
    df['train_acc'] = rnd_df.train_acc/metisp1_df.train_acc
    df['test_loss'] = rnd_df.test_loss/metisp1_df.test_loss
    df['test_acc'] = rnd_df.test_acc/metisp1_df.test_acc
    noniidL20E5_dfs.append(df)

```

```python
line_color = ["#0e6787", "#ad85ab", "#f0bca5"]
#line_color1 = ["#021c1e", "#2c7873", "#6fb98f"]
line_color1=["#db9501","#c05805","#2e2300"]
#ff706b
#ffa600
line_color2 = ["#2d4262", "#73605b", "#d09683"]
line_color3 = ["#bcbd22", "#f56c57", "#17becf"] #"#7f7f7f"
#line_color3 = ["#f4cc70","#07575b", "#20948b"]

line_color2 = [ 'mediumblue', 'orangered','purple']

lbl_metisp1_cluster = ["eFL_metis: #edge_servers = 10", "eFL_metis: #edge_servers = 20", "eFL_metis: #edge_servers = 30"]
lbl_rnd_cluster = ["eFL_rnd: #edge_servers = 10", "eFL_rnd: #edge_servers = 20", "eFL_rnd: #edge_servers = 30"]
lbl_cluster = ["#servers = 10", "#servers = 20", "#servers = 30"]
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.figure.set_dpi(200)
for i in range(noniidL10E2_n_files):
    df = noniidL10E2_dfs[i][noniidL10E2_dfs[i].global_round <= 3000]
    ax.plot(df.global_round, df.test_acc, label = lbl_cluster[i], c = line_color2[i], linewidth = 1)
ax.set_title("CIFAR10: zipf, L=10, E=2, eFL_new vs. eFL_rnd", size = fontsize/1.4)
ax.set_ylabel("Test Accuracy Ratio", size=fontsize)
ax.set_xlabel("Global Time Steps", size=fontsize)
ax.legend(loc='best', fontsize=fontsize/1.3)
plt.xticks(fontsize = fontsize/1.3)
plt.yticks(fontsize = fontsize/1.3)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.figure.set_dpi(200)
for i in range(noniidL20E5_n_files):
    df = noniidL20E5_dfs[i][noniidL20E5_dfs[i].global_round <= 1500]
    ax.plot(df.global_round, df.test_acc, label = lbl_cluster[i], c = line_color2[i], linewidth = 1)
ax.set_title("CIFAR10: zipf, L=20, E=5, eFL_new vs. eFL_rnd", size = fontsize/1.4)
ax.set_ylabel("Test Accuracy Ratio", size=fontsize)
ax.set_xlabel("Global Time Steps", size=fontsize)
ax.legend(loc='best', fontsize = fontsize/1.3)
plt.xticks(fontsize = fontsize/1.3)
plt.yticks(fontsize = fontsize/1.3)
plt.show()
```
