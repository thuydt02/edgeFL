```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
iidL10E2_path_file = "./../output/mnist/acc/iid/L10E2/"
iidL20E5_path_file = "./../output/mnist/acc/iid/L20E5/"
noniidL10E2_path_file = "./../output/mnist/acc/noniid/L10E2/"
noniidL20E5_path_file = "./../output/mnist/acc/noniid/L20E5/"
fontsize = 20
```

```python
#reading files

#div-label1
noniidL10E2_metisp1_files = [
    "eFL_LRlr0.01_dc1.0_B10_L10_E2_G3000_g_nw_d_minkowski_p1.0_MLP2_G10_partition_noniid90_nclient300.npy.part.30.weight.pth.acc.csv"
]
#rnd-label1
noniidL10E2_rnd_files = [
    "eFL_LRlr0.01_dc1.0_B10_L10_E2_G3000_1z_rnd_num_client300.part30_partition_noniid90_nclient300.npy.weight.pth.acc.csv"
]
#div-lable1
noniidL20E5_metisp1_files = [
    "eFL_LRlr0.01_dc1.0_B10_L20_E5_G3000_g_nw_d_minkowski_p1.0_MLP2_G10_partition_noniid90_nclient300.npy.part.30.weight.pth.acc.csv"
]
#rnd-label1
noniidL20E5_rnd_files = [
   "eFL_LRlr0.01_dc1.0_B10_L20_E5_G3000_1z_rnd_num_client300.part30_partition_noniid90_nclient300.npy.weight.pth.acc.csv"
]

iidL10E2_metisp1_files = [
    "eFL_MLP2lr0.01_dc1.0_B10_L10_E2_G3000_g_nw_d_minkowski_p1.0_MLP2_G10_partition_iid_nclient300.npy.part.10.weight.pth.acc.csv",
    "eFL_MLP2lr0.01_dc1.0_B10_L10_E2_G3000_g_nw_d_minkowski_p1.0_MLP2_G10_partition_iid_nclient300.npy.part.20.weight.pth.acc.csv",
    "eFL_MLP2lr0.01_dc1.0_B10_L10_E2_G3000_g_nw_d_minkowski_p1.0_MLP2_G10_partition_iid_nclient300.npy.part.30.weight.pth.acc.csv"
]

iidL10E2_rnd_files = [
    "eFL_MLP2lr0.01_dc1.0_B10_L10_E2_G3000_avg10z_rnd_num_client300.part10_partition_iid_nclient300.npy.weight.pth.acc.csv",
    "eFL_MLP2lr0.01_dc1.0_B10_L10_E2_G3000_avg10z_rnd_num_client300.part20_partition_iid_nclient300.npy.weight.pth.acc.csv",
    "eFL_MLP2lr0.01_dc1.0_B10_L10_E2_G3000_avg10z_rnd_num_client300.part30_partition_iid_nclient300.npy.weight.pth.acc.csv"
]

iidL20E5_metisp1_files = [
    "eFL_MLP2lr0.01_dc1.0_B10_L20_E5_G5000_g_nw_d_minkowski_p1.0_MLP2_G10_partition_iid_nclient300.npy.part.10.weight.pth.acc.csv",
    "eFL_MLP2lr0.01_dc1.0_B10_L20_E5_G5000_g_nw_d_minkowski_p1.0_MLP2_G10_partition_iid_nclient300.npy.part.20.weight.pth.acc.csv",
    "eFL_MLP2lr0.01_dc1.0_B10_L20_E5_G5000_g_nw_d_minkowski_p1.0_MLP2_G10_partition_iid_nclient300.npy.part.30.weight.pth.acc.csv"
]

iidL20E5_rnd_files = [
    "eFL_MLP2lr0.01_dc1.0_B10_L20_E5_G5000_avg10z_rnd_num_client300.part10_partition_iid_nclient300.npy.weight.pth.acc.csv",
    "eFL_MLP2lr0.01_dc1.0_B10_L20_E5_G5000_avg10z_rnd_num_client300.part20_partition_iid_nclient300.npy.weight.pth.acc.csv",
    "eFL_MLP2lr0.01_dc1.0_B10_L20_E5_G5000_avg10z_rnd_num_client300.part30_partition_iid_nclient300.npy.weight.pth.acc.csv"
]



noniidL10E2_dfs = []
noniidL10E2_n_files = len(noniidL10E2_metisp1_files)
for i in range(noniidL10E2_n_files):
    metisp1_df = pd.read_csv(noniidL10E2_path_file + noniidL10E2_metisp1_files[i])
    rnd_df = pd.read_csv(noniidL10E2_path_file + noniidL10E2_rnd_files[i])
    df = pd.DataFrame(metisp1_df.global_round)
    
    df['train_loss'] = rnd_df.train_loss / metisp1_df.train_loss
    df['train_acc'] = rnd_df.train_acc / metisp1_df.train_acc
    df['test_loss'] = rnd_df.test_loss / metisp1_df.test_loss 
    df['test_acc'] = rnd_df.test_acc / metisp1_df.test_acc

    noniidL10E2_dfs.append(df)

noniidL20E5_dfs = []
noniidL20E5_n_files = len(noniidL20E5_metisp1_files)
for i in range(noniidL20E5_n_files):
    metisp1_df = pd.read_csv(noniidL20E5_path_file + noniidL20E5_metisp1_files[i])
    rnd_df = pd.read_csv(noniidL20E5_path_file + noniidL20E5_rnd_files[i])
    df = pd.DataFrame(metisp1_df.global_round)
    df['train_loss'] = rnd_df.train_loss / metisp1_df.train_loss
    df['train_acc'] = rnd_df.train_acc / metisp1_df.train_acc
    df['test_loss'] = rnd_df.test_loss / metisp1_df.test_loss 
    df['test_acc'] = rnd_df.test_acc / metisp1_df.test_acc
    noniidL20E5_dfs.append(df)

iidL10E2_dfs = []
iidL10E2_n_files = len(iidL10E2_metisp1_files)
for i in range(iidL10E2_n_files):
    metisp1_df = pd.read_csv(iidL10E2_path_file + iidL10E2_metisp1_files[i])
    rnd_df = pd.read_csv(iidL10E2_path_file + iidL10E2_rnd_files[i])
    df = pd.DataFrame(metisp1_df.global_round)
    df['train_loss'] = rnd_df.train_loss / metisp1_df.train_loss
    df['train_acc'] = rnd_df.train_acc / metisp1_df.train_acc
    df['test_loss'] = rnd_df.test_loss / metisp1_df.test_loss 
    df['test_acc'] = rnd_df.test_acc / metisp1_df.test_acc
    iidL10E2_dfs.append(df)

iidL20E5_dfs = []
iidL20E5_n_files = len(iidL20E5_metisp1_files)
for i in range(iidL20E5_n_files):
    metisp1_df = pd.read_csv(iidL20E5_path_file + iidL20E5_metisp1_files[i])
    rnd_df = pd.read_csv(iidL20E5_path_file + iidL20E5_rnd_files[i])
    df = pd.DataFrame(metisp1_df.global_round)
    df['train_loss'] = metisp1_df.train_loss / rnd_df.train_loss
    df['train_acc'] = metisp1_df.train_acc / rnd_df.train_acc
    df['test_loss'] = metisp1_df.test_loss / rnd_df.test_loss
    df['test_acc'] = metisp1_df.test_acc / rnd_df.test_acc
    iidL20E5_dfs.append(df)

```

```python
#plot
line_color = ["#0e6787", "#ad85ab", "#f0bca5"]
line_color2 = [ 'mediumblue', 'orangered','purple']

lbl_metisp1_cluster = ["eFL_metis: #edge_servers = 10", "eFL_metis: #edge_servers = 20", "eFL_metis: #edge_servers = 30"]
lbl_rnd_cluster = ["eFL_rnd: #edge_servers = 10", "eFL_rnd: #edge_servers = 20", "eFL_rnd: #edge_servers = 30"]
lbl_cluster = ["#servers = 10", "#servers = 20", "#servers = 30"]
fig, ax = plt.subplots(2, 2, figsize=(15,10))

for i in range(iidL10E2_n_files):
    df = iidL10E2_dfs[i][iidL10E2_dfs[i].global_round <= 5000]
    ax[0,0].plot(df.global_round, df.test_acc, label = lbl_cluster[i], c = line_color[i], linewidth = 3)
ax[0,0].set_title("MNIST: iid, L=10, E=2")
ax[0,0].set_ylabel("Test Accuracy Ratio of eFL_metis to eFL_rnd")
ax[0,0].set_xlabel("Global Time Steps")
ax[0,0].legend(loc='best')

for i in range(iidL20E5_n_files):
    df = iidL20E5_dfs[i][iidL20E5_dfs[i].global_round <= 5000]
    ax[0,1].plot(df.global_round, df.test_acc, label = lbl_cluster[i], c = line_color[i], linewidth = 3)
ax[0,1].set_title("MNIST: iid, L=20, E=5")
ax[0,1].set_ylabel("Test Accuracy Ratio of eFL_metis to eFL_rnd")
ax[0,1].set_xlabel("Global Time Steps")
ax[0,1].legend(loc='best')

for i in range(noniidL10E2_n_files):
    df = noniidL10E2_dfs[i][noniidL10E2_dfs[i].global_round <= 5000]
    ax[1,0].plot(df.global_round, df.test_acc, label = lbl_cluster[i], c = line_color[i], linewidth = 3)
ax[1,0].set_title("MNIST: noniid, L=10, E=2")
ax[1,0].set_ylabel("Test Accuracy Ratio of eFL_metis to eFL_rnd")
ax[1,0].set_xlabel("Global Time Steps")
ax[1,0].legend(loc='best')

for i in range(noniidL20E5_n_files):
    df = noniidL20E5_dfs[i][noniidL20E5_dfs[i].global_round <= 5000]
    ax[1,1].plot(df.global_round, df.test_acc, label = lbl_cluster[i], c = line_color[i], linewidth = 3)
ax[1,1].set_title("MNIST: noniid, L=20, E=5")
ax[1,1].set_ylabel("Test Accuracy Ratio of eFL_metis to eFL_rnd")
ax[1,1].set_xlabel("Global Time Steps")
ax[1,1].legend(loc='best')

plt.show()
```
