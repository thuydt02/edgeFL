```python
import random
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(15,13))
miu, sigma = 0, .25 #-1, 0.5
n = 300
s = np.random.lognormal(miu, sigma, n)
ax.hist(s, bins = 50)
plt.show()
```

```python
path_file = "./output/mnist/z_ass/"
fname = "q_lognorm_part" + str(n) + ".npy"
for i in range (len(s)):
    if (s[i] > 1) or (s[i] < 0.05):
        s[i] = random.uniform(0,1)
with open(path_file + fname, 'wb') as the_file:
    np.save(the_file, s)
```

```python
fig, ax = plt.subplots(figsize=(15,13))
ax.hist(s, bins = 10)
plt.show()
```
