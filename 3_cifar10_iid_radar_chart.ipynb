```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
path_file_L10E2 = "./output/cifar10/acc/iid/L10E2/"
path_file_L20E5 = "./output/cifar10/acc/iid/L20E5/"
fontsize = 20
```

```python
#reading files
metisp1_files_L10E2 = [
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G3000_g_nw_d_minkowski_p1.0_CNN2_G100_partition_iid_nclient300.npy.part.10.weight.pth.acc.csv",
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G3000_g_nw_d_minkowski_p1.0_CNN2_G100_partition_iid_nclient300.npy.part.20.weight.pth.acc.csv",
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G3000_g_nw_d_minkowski_p1.0_CNN2_G100_partition_iid_nclient300.npy.part.30.weight.pth.acc.csv"
]

metisp1_files_L20E5 = [
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L20_E5_G3000_g_nw_d_minkowski_p1.0_CNN2_G100_partition_iid_nclient300.npy.part.10.weight.pth.acc.csv",
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L20_E5_G3000_g_nw_d_minkowski_p1.0_CNN2_G100_partition_iid_nclient300.npy.part.20.weight.pth.acc.csv",
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L20_E5_G3000_g_nw_d_minkowski_p1.0_CNN2_G100_partition_iid_nclient300.npy.part.30.weight.pth.acc.csv"
]

rnd_files_L10E2 = [
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G5000_avg5z_rnd_num_client300.part10_partition_iid_nclient300.npy.weight.pth.acc.csv",
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G5000_avg10z_rnd_num_client300.part20_partition_iid_nclient300.npy.weight.pth.acc.csv",
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L10_E2_G5000_avg10z_rnd_num_client300.part30_partition_iid_nclient300.npy.weight.pth.acc.csv"
]

rnd_files_L20E5 = [
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L20_E5_G3000_avg5z_rnd_num_client300.part10_partition_iid_nclient300.npy.weight.pth.acc.csv",
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L20_E5_G3000_avg10z_rnd_num_client300.part20_partition_iid_nclient300.npy.weight.pth.acc.csv",
    "eFL_CNN2_dr10lr0.15_dc0.99_B20_L20_E5_G3000_avg10z_rnd_num_client300.part30_partition_iid_nclient300.npy.weight.pth.acc.csv"
]





#metisp1_dfs_L10E2 = []
#metisp1_dfs_L20E5 = []
#rnd_dfs_L10E2 = []
#rnd_dfs_L20E5 = []

n_metisp1_files_L10E2 = len(metisp1_files_L10E2)
n_metisp1_files_L20E5 = len(metisp1_files_L20E5)
n_rnd_files_L10E2 = len(rnd_files_L10E2)
n_rnd_files_L20E5 = len(rnd_files_L20E5)

metisp1_max_test_acc_L10E2 = []
metisp1_max_test_acc_L20E5 = []
rnd_max_test_acc_L10E2 = []
rnd_max_test_acc_L20E5 = []

for i in range(n_metisp1_files_L10E2):
    #metisp1_dfs_L10E2.append(pd.read_csv(path_file_L10E2 + metisp1_files_L10E2[i]))
    df  = pd.read_csv(path_file_L10E2 + metisp1_files_L10E2[i])
    metisp1_max_test_acc_L10E2.append(df.tail(1).test_acc.values[0])

for i in range(n_metisp1_files_L20E5):
    #metisp1_dfs_L10E2.append(pd.read_csv(path_file_L20E5 + metisp1_files_L20E5[i]))
    df  = pd.read_csv(path_file_L20E5 + metisp1_files_L20E5[i])
    metisp1_max_test_acc_L20E5.append(df.tail(1).test_acc.values[0])

for i in range(n_rnd_files_L10E2):
    #rnd_dfs.append(pd.read_csv(path_file_L10E2 + rnd_files_L10E2[i]))
    df  = pd.read_csv(path_file_L10E2 + rnd_files_L10E2[i])
    rnd_max_test_acc_L10E2.append(df.tail(1).test_acc.values[0])

for i in range(n_rnd_files_L20E5):
    #rnd_dfs.append(pd.read_csv(path_file_L20E5 + rnd_files_L20E5[i]))
    df  = pd.read_csv(path_file_L20E5 + rnd_files_L20E5[i])
    rnd_max_test_acc_L20E5.append(df.tail(1).test_acc.values[0])



```
