```python
import numpy as np
```

```python
def draw_zipf_distribution(n_clients, z):
    p = (1/np.arange(1,n_clients + 1))**(z)
    return p/sum(p)
```

```python
import pandas as pd

n_clients = 300
z = 0.5
path_file = "./output/cifar10/z_ass/"
data_size_file = "data_size_zipfz" + str(z) + "_nclient" + str(n_clients) + ".csv"
p = draw_zipf_distribution(n_clients, z)

df = pd.DataFrame(data = p)
df.to_csv(path_file + data_size_file, header = None, index = None)
```

```python
from matplotlib import pyplot as plt
plt.plot(range(len(p)),p)
plt.show()
```
